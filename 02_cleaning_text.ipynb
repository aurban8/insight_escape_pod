## Use NLTK to tokenize words

```python
from __future__ import division  # Python 2 users only
import nltk, re, pprint
from nltk import word_tokenize
```

## Remove stop words

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords[:5]
```

```python
stop_remover = lambda x: [w for w in x if w.lower() not in stopwords]    
```

```python
escape_pod['punct_gone'][9][:10]
```

```python
stop_remover(escape_pod['punct_gone'][9])[:10]
```

```python
escape_pod['stop_gone'] = escape_pod['punct_gone'].apply(stop_remover)
```

```python
escape_pod['stop_gone']
```

## Stemmer

```python
# stemmers
porter = nltk.PorterStemmer()
```

```python
port_stemmer = lambda x: [porter.stem(item) for item in x]
```

```python
escape_pod['stem'] = escape_pod['stop_gone'].apply(port_stemmer)
```

## Some stats on stop word, punct removed, stem words

### Top words in all words

```python
import nltk
```

```python
all_word_list = escape_pod['stem'].tolist()
all_word_list = [val for sublist in all_word_list for val in sublist]
```

```python
fdist = nltk.FreqDist(all_word_list)
```

```python
fdist.most_common(10)
```

#### Find top 10 words for each podcast

```python
top10func = lambda x: [(nltk.FreqDist(x)).most_common(10) ]
```

```python
escape_pod['top10'] = escape_pod['stem'].apply(top10func)
```
