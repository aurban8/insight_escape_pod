```python
def article_list(url):
###   Function extracts all Escape Pod permanent articles links from a given website 
###   Input: website URL
###   Output: list of links to articles 
    response = get(url)
    html = response.content
    soup = BeautifulSoup(html,'lxml')
#    print soup.prettify()[0:500]  # print first few characters
# Store all results of blog posts in perm_links_tags
    perm_link_tags = soup.find_all('div', class_ = 'entry')
#    print(type(perm_link_tags))
#    print(len(perm_link_tags))
# View first few characters of first entry
#    print perm_links[0].prettify()[0:1000]
# Access episode title
#    perm_link_tags[0].h2.a.text
#    perm_link_tags[0].a
# Access blog post direct link.
#    perm_link_tags[9].h2.a['href']
#    print len(perm_link_tags)
#    for i in range(len(perm_link_tags)):
#        print i, perm_link_tags[i]
    perm_links = []

    for tag in perm_link_tags:
#        print tag.h2
        if tag.find("h2"):
#            print tag.h2.a['href']
            perm_links.append(tag.h2.a['href'] )


#    
#    for i in range(len(perm_link_tags)):
#       if perm_link_tags[i].h2.a['href']:
#            perm_links+= perm_link_tags[i].h2.a['href']
#    print perm_links[0]
    return perm_links
```

```python
def get_post_data(html):
# input: html data created by function 'create_html_object'
# output --> post_data: dictionary with post data 
# items in post_data dictionary: author, narrator, host,...
# ... tags, filed in, rating,  rated tags, number of comments

    soup = BeautifulSoup(html,'lxml')
    post_data = []    # output dictionary 
#    post_info = soup.find('div', class_ = 'postinfo')
#    print post_info
#    print post_info.prettify()[0:500] 

# Find post category info
#    post_info = soup.find_all('a', rel="category tag")
# Find post_title
    post_title = soup.find('h1', class_ = 'post-title')
    post_data.append(post_title.text)  # post_data.append(str(post_title.text))
# Find post info which contains filed tags and number of comments. 
    post_info = soup.find('div', class_ = 'postinfo')
# Find find filed tags within post_info
    filed_tags = []
    if post_info.find_all('a', rel = 'category tag'):
        tags = post_info.find_all('a', rel = 'category tag')
        for tag in tags:           
            filed_tags.append(tag.text)  #            filed_tags.append(str(tag.text))
    post_data.append(filed_tags)
# Find number of comments within post_info
    comments = post_info.find_all('a')
    comments = comments[-1].text
    post_data.append(comments)   #    post_data.append(str(comments))
# Find side tags
    side_tags = []
    if soup.find('li', class_ ='tags'):
        tags = soup.find('li', class_ = 'tags')
        tags1 = tags.find_all('a', rel = 'tag')
        for tag in tags1:
            side_tags.append(tag.text)   #            side_tags.append(str(tag.text))
    post_data.append(side_tags)
# Find podcast runtime/duration
    duration = None
    if soup.find('span', class_ = "podpress_mediafile_dursize podpress_mediafile_dursize_audio_mp3"):
        duration =  soup.find('span', class_ = "podpress_mediafile_dursize podpress_mediafile_dursize_audio_mp3")
        duration = duration.text    #        duration = str(duration.text)
    post_data.append(duration)

# Find story text, author, narrator, host
    story = []
    story_snippet = []
    author = None
    host = None
    narrator = None
    author_info = None
    collect_rest = False  # if True, collect all remaining fields in p tag as story text.
    post_type = 0
    zeroth_line = ''

    
# post_type key
# 99 = NOT A PODCAST, Title doesn't start with "EP###:"
# 0 = unknown 
# 1 = AUTHOR, NARRATOR, HOST, author, narrator info, story. ex: http://escapepod.org/2017/05/11/ep575-red-kelly-owns-the-moon/
# 2 = By Author (uses b tag), story snippet, story rating and some tags. ex: http://escapepod.org/2006/11/23/ep081-margin-of-error/
# 3 = By Author, Read by, story, ex: http://escapepod.org/2011/09/30/ep312_night_bird_soaring/
# 4 = By Author (uses strong tag), story snippet, ex: http://escapepod.org/2005/05/12/ep001-imperial-by-jonathon-sullivan/

# Check that post's title format is EP###: 
# Then it's a podcast story episode.  
# Now collect story data. 
    if post_title.text[0:2]=="EP" and post_title.text[5:6] == ":":
        post = soup.find('div', class_ = 'entry')
        if post.find_all('p'):
            for content in post.find_all('p'):
                if collect_rest: 
                    story.append(content.text)
                if post_type == 1 and content.text[0:4] == zeroth_line:
                    collect_rest = True
# POST TYPE = 1 
                if content.text[0:7] == 'AUTHOR:':
                    author = content.text[8:]
                    post_type = 1
                    zeroth_line = 'By ' + author[0]
                # Find text that matches "By " + author's first initial 
                # (Can't use full name because not always the same)
                # ex: http://escapepod.org/2017/05/11/ep575-red-kelly-owns-the-moon/
                if content.text[0:9] == 'NARRATOR:':
                    narrator = content.text[10:]
                if content.text[0:5] == 'HOST:':
                    host = content.text[6:]
# POST TYPE = 2
                if content.find('b'):
                    if content.find('b').text[0:2] == 'By':
                        author = content.find('b').text[3:]
                        post_type = 2
                        if author[-1] == '.':
                            author = author[:-1]
                        collect_rest = True
#                print post_type
# POST TYPE = 3, 4
                if content.find('strong'):
                    if content.find('strong').text[0:2] == 'By':
                        author = content.find('strong').text[3:]
                        title_length = len(content.find('strong').text) * -1
                        post_type = 3  # assume type 3, unless find italics.
                        print 'set posttype =3'
#                    print content.find('strong').text, 'hi', post_title.text[title_length:]

                    if post_type == 3 and content.find('strong').text == post_title.text[title_length:]:
                        collect_rest = True
                if post_type == 3 and content.find('i'):
                    collect_rest = True
                    post_type = 4
                    story.append(content.text)
                if post_type == 3 and content.find('em'):
                    print post_type,'hi'
                    collect_rest = True
                    post_type = 4
                    print post_type,'hi'
                    story.append(content.text)
        print post_type,'hi'
#                print zeroth_line
    else:
        post_type = 99  # NOT A PODCAST
    
# remove last two lines of story for post_type = 2 and move to story_snippet
    if post_type == 2:
        story_snippet = story[:-2]
        story = []

# change story stored for type 4 
    if post_type == 4:
        story_snippet = story[:-1]
        story = []


    post_data.append(author)
    post_data.append(narrator)
    post_data.append(host)
    post_data.append(post_type)
    post_data.append(story_snippet)
    post_data.append(story)
# Summary of what this function outputs:    
# post_data is a list that contains: 
# post's title,
# list of filed_tags
# number of comments
# list of side tags
# podcast runtime/duration
# author
# narrator
# host
# post_type 0 = unknown, 1 = classic, 2 = old skool
# story snippet
# story content

    return post_data
```

```python
def get_post_data_ptags(html):
# input: html data created by function 'create_html_object'
# output --> post_data: dictionary with post data 
# items in post_data dictionary: author, narrator, host,...
# ... tags, filed in, rating,  rated tags, number of comments

    soup = BeautifulSoup(html,'lxml')
    post_data = []    # output dictionary 
#    post_info = soup.find('div', class_ = 'postinfo')
#    print post_info
#    print post_info.prettify()[0:500] 

# Find post category info
#    post_info = soup.find_all('a', rel="category tag")
# Find post_title
    post_title = soup.find('h1', class_ = 'post-title')
    post_data.append(post_title.text)  # post_data.append(str(post_title.text))
# Find post info which contains filed tags and number of comments. 
    post_info = soup.find('div', class_ = 'postinfo')
# Find find filed tags within post_info
    filed_tags = []
    if post_info.find_all('a', rel = 'category tag'):
        tags = post_info.find_all('a', rel = 'category tag')
        for tag in tags:           
            filed_tags.append(tag.text)  #            filed_tags.append(str(tag.text))
    post_data.append(filed_tags)
# Find number of comments within post_info
    comments = post_info.find_all('a')
    comments = comments[-1].text
    post_data.append(comments)   #    post_data.append(str(comments))
# Find side tags
    side_tags = []
    if soup.find('li', class_ ='tags'):
        tags = soup.find('li', class_ = 'tags')
        tags1 = tags.find_all('a', rel = 'tag')
        for tag in tags1:
            side_tags.append(tag.text)   #            side_tags.append(str(tag.text))
    post_data.append(side_tags)
# Find podcast runtime/duration
    duration = None
    if soup.find('span', class_ = "podpress_mediafile_dursize podpress_mediafile_dursize_audio_mp3"):
        duration =  soup.find('span', class_ = "podpress_mediafile_dursize podpress_mediafile_dursize_audio_mp3")
        duration = duration.text    #        duration = str(duration.text)
    post_data.append(duration)

# Find story text, author, narrator, host
    story = []
    author = None
    host = None
    narrator = None
    author_info = None
    collect_rest = True  # if True, collect all remaining fields in p tag as story text.
    post_type = 0
    zeroth_line = ''

    
# post_type key
# 99 = NOT A PODCAST, Title doesn't start with "EP###:"
# 0 = unknown 
# 1 = AUTHOR, NARRATOR, HOST, author, narrator info, story. ex: http://escapepod.org/2017/05/11/ep575-red-kelly-owns-the-moon/
# 2 = By Author (uses b tag), story snippet, story rating and some tags. ex: http://escapepod.org/2006/11/23/ep081-margin-of-error/
# 3 = By Author, Read by, story, ex: http://escapepod.org/2011/09/30/ep312_night_bird_soaring/
# 4 = By Author (uses strong tag), story snippet, ex: http://escapepod.org/2005/05/12/ep001-imperial-by-jonathon-sullivan/

# Check that post's title format is EP###: 
# Then it's a podcast story episode.  
# Now collect story data. 
    if post_title.text[0:2]=="EP" and post_title.text[5:6] == ":":
        post = soup.find('div', class_ = 'entry')
        if post.find_all('p'):
            for content in post.find_all('p'):
                if collect_rest: 
                    story.append(content.text)
    else:
        post_type = 99  # NOT A PODCAST
    

    post_data.append(author)
    post_data.append(narrator)
    post_data.append(host)
    post_data.append(post_type)
    post_data.append(story)
# Summary of what this function outputs:    
# post_data is a list that contains: 
# post's title,
# list of filed_tags
# number of comments
# list of side tags
# podcast runtime/duration
# author
# narrator
# host
# post_type 0 = unknown, 1 = classic, 2 = old skool
# story snippet
# story content
    post_data.append(" ".join(story))
    return post_data
```
